```python
import pandas as pd

var = pd.read_csv("C:\\Users\\Admin\\OneDrive\\Desktop\\Delinquency1.csv")


missing_summary = var.isnull().sum()
print("Missing Values Summary:\n", missing_summary)


numerical_cols = var.select_dtypes(include=['int64', 'float64']).columns

for col in numerical_cols:
    if var[col].isnull().sum() > 0:
        median_val = var[col].median()
        var[col].fillna(median_val, inplace=True)
        print(f"Filled missing values in '{col}' with median: {median_val}")


categorical_cols = var.select_dtypes(include=['object']).columns

for col in categorical_cols:
    if var[col].isnull().sum() > 0:
        mode_val = var[col].mode()[0]
        var[col].fillna(mode_val, inplace=True)
        print(f"Filled missing values in '{col}' with mode: {mode_val}")


print("\nMissing values after treatment:\n", var.isnull().sum())
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.heatmap(var.corr(), annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()

sns.boxplot(x='Delinquency_Flag', y='Credit_Utilization', data=var)
plt.title("Credit Utilization by Delinquency")
plt.show()
```

```python
num_cols = var.select_dtypes(include='number').columns
target_col = "Delinquent_Account" 

corrs = var[num_cols].corrwith(var[target_col]).sort_values(key=lambda s: s.abs(), ascending=False)
print(corrs)
```
